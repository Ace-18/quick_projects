### Given the names and grades for each student in a class of  students, store them in a nested list and print the name(s) of any student(s) having the second lowest grade.

Note: If there are multiple students with the second lowest grade, order their names alphabetically and print each name on a new line.

Input Format

The first line contains an integer, <b>N</b>  , the number of students.
The <b>2 N</b> subsequent lines describe each student over  lines.
- The first line contains a student's name.
- The second line contains their grade.

```python
records=[]
for _ in range(int(input())):
    name = input()
    score = float(input())
    records.append([name,score])
```

```python
records
```

```python
#records.sort(key=lambda record: record[1])

s=list(set([record[1] for record in records]))
s.sort()

s_names=[record[0] for record in records if record[1]==s[1]]

s_names.sort()
```

```python
print(*s_names, sep = "\n")
```

### The provided code stub will read in a dictionary containing key/value pairs of name:[marks] for a list of students. Print the average of the marks array for the student name provided, showing 2 places after the decimal.



#### The query_name is 'beta'. beta's average score is (30+50+70)/3 = 50.0

### Input Format

#### The first line contains the integer , the number of students' records. The next  lines contain the names and marks obtained by a student, each value separated by a space. The final line contains query_name, the name of a student to query.

```python
n = int(input())
student_marks = {}
for _ in range(n):
    name, *line = input().split()
    scores = list(map(float, line))
    student_marks[name] = scores
query_name = input()
```

```python
3
Krishna 67 68 69
Arjun 70 98 63
Malika 52 56 60
Malika
```

```python
student_average=sum(student_marks[query_name])/len(student_marks[query_name])
print('{:.2f}'.format(student_average))
```

```python
print('%.2f' % student_average)
```

### eye

The eye tool returns a 2-D array with 1's as the diagonal and 0's elsewhere. The diagonal can be main, upper or lower depending on the optional parameter <b>k</b>. A positive <b>k</b> is for the upper diagonal, a negative <b>k</b> is for the lower, and a <b>0k</b>(default) is for the main diagonal.

import numpy<br>
print(numpy.eye(8, 7, k = 1))    # 8 X 7 Dimensional array with first upper diagonal 1.<br>

[[ 0.  1.  0.  0.  0.  0.  0.]<br>
